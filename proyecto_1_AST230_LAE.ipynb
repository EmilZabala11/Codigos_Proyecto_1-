```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = 6.672e-11 
M = 1.4e12 * 1.989e30

r = []
V = []
n = 1

for i in range(100):
    r0 = n*3.086e19
    r.append(r0)
    V.append(np.sqrt((G*M)/r0))
    n = n+1

plt.plot(r,V)
plt.xlabel("Radio [m]")
plt.ylabel("Vcirc [m/s]")
plt.show()
```
